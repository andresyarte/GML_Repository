```python
df.columns = list(map(str, df.columns))
df.columns
```

```python
df.set_index('Country', inplace = True)
df.head()
```

```python
df['Total'] = df.iloc[:,
                      3:].sum(axis=1)
df.head(2)
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

#pip install matplotlib
```

```python
mpl.style.use('ggplot')
```

```python
print(plt.style.available)
```

```python
j_data = df.loc['Japan','1980':'2013']
j_data.plot()
```

```python
ic_data = df.loc[['India','China'],'1980':'2013']
```

```python
ic_data
```

```python
nic = ic_data.transpose()
```

```python
nic.plot(figsize = (20,10))
plt.xlabel('Years', fontsize = 17)
plt.xticks(fontsize = 17, rotation = 45)
plt.ylabel('Number of imigrants', fontsize = 17)
plt.yticks(fontsize = 17)
plt.title('Immigration to canada form India and China during 1980 to 2013', fontsize = 17)
plt.show()
```
